# Introduction

**NLP spam text classification task with pretrained transformer model**

Model official page in huggingface: [FredZhang7/one-for-all-toxicity-v3](https://huggingface.co/FredZhang7/one-for-all-toxicity-v3)

# Important libraries and cuda initializing

```python
import torch

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print('device:', device)
```

# Loading Tokenizer

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("FredZhang7/one-for-all-toxicity-v3")
```

# Loading Model

```python
from transformers import AutoModelForSequenceClassification

model = AutoModelForSequenceClassification.from_pretrained("FredZhang7/one-for-all-toxicity-v3").to(device)
```

# Prediction

Toxic text example

```python
inputs = "مردک بیمار"

encoding = tokenizer.encode_plus(
    inputs,
    add_special_tokens=True,
    max_length=208,
    padding="max_length",
    truncation=True,
    return_tensors="pt"
)
input_ids = encoding["input_ids"].to(device)
attention_mask = encoding["attention_mask"].to(device)

with torch.no_grad():
    outputs = model(input_ids, attention_mask=attention_mask)
    logits = outputs.logits
    predicted_labels = torch.argmax(logits, dim=1)

print(predicted_labels)
```

non-toxic text example

```python
inputs = "دوست دارم!"

encoding = tokenizer.encode_plus(
    inputs,
    add_special_tokens=True,
    max_length=208,
    padding="max_length",
    truncation=True,
    return_tensors="pt"
)
input_ids = encoding["input_ids"].to(device)
attention_mask = encoding["attention_mask"].to(device)

with torch.no_grad():
    outputs = model(input_ids, attention_mask=attention_mask)
    logits = outputs.logits
    predicted_labels = torch.argmax(logits, dim=1)

print(predicted_labels)
```
